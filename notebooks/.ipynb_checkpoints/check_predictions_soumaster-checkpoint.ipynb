# Setup

```python
current_directory = os.getcwd()
# Volta uma pasta (vai para o diretório pai)
parent_directory = os.path.dirname(current_directory)

# Cria uma pasta chamada 'datasets' no diretório pai
datasets_directory = os.path.join(parent_directory, 'datasets')

INPUT_PATH_VALIDATION = f'{datasets_directory}\\dataset_A3Data_validation.csv'
MODEL_NAME = 'churn.pkl'
PATH_OUTPUT = datasets_directory
model_features =  ['SeniorCitizen',
                  'Partner',
                  'Dependents',
                  'tenure',
                  'PhoneService',
                  'MultipleLines',
                  'InternetServiceFiber',
                  'OnlineSecurity',
                  'OnlineBackup',
                  'DeviceProtection',
                  'TechSupport',
                  'StreamingTV',
                  'StreamingMovies',
                  'PaperlessBilling',
                  'MonthlyCharges',
                  'TotalCharges',
                  'Churn',
                  'gender_Male',
                  'Contract_One year',
                  'Contract_Two year',
                  'PaymentMethod_Credit card (automatic)',
                  'PaymentMethod_Electronic check',
                  'PaymentMethod_Mailed check'
                 ] 

target_name = "Churn"
```

# Load model

```python
with open(MODEL_NAME, 'rb') as model:
    model_xgboost = pickle.load(model)
```

# Reading validation dataset

```python
try:
    del df_validation
    del df_validation_model 
except:
    pass

df_validation = pd.read_csv(INPUT_PATH_VALIDATION)
df_validation_model = df_validation[model_features]
```

# Create dummies

```python
cat_features = list(df_validation_model.drop('Churn', axis=1).select_dtypes(include=['object','bool']).columns)
num_features = list(df_validation_model.select_dtypes(include=['int32','int64','float64','float32']).columns)
```

```python
for cat_feature in cat_features:
    dummies = pd.get_dummies(df_validation_model[cat_feature], prefix="var_" + cat_feature)
    df_validation_model = pd.concat([df_validation_model.drop(cat_feature, axis=1), dummies], axis=1)

regex = re.compile(r"\[|\]|<", re.IGNORECASE)

df_validation_model.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in df_validation_model.columns.values]
```

```python
list(df_validation_model.columns)
```

# Generate predictions

```python
df_validation_model = df_validation_model.dropna()
X_validation =df_validation_model.drop(target_name, axis=1)
y_validation = df_validation_model[target_name]
```

```python
y_pred_validation = model_xgboost.predict(X_validation)
y_prob_validation = model_xgboost.predict_proba(X_validation)
print('Accuracy of classification model on validation set: {:.2f}'.format(model_xgboost.score(X_validation, y_validation)))
```

```python
df_pred_validation = pd.DataFrame({'Index': y_validation.index, 'Actual': y_validation.ravel(), 'Predicted': y_pred_validation.ravel(), 'Score' : (1000*y_prob_validation[:,0]).ravel()})
```

```python
df_pred_validation.sort_values(by=['Index'], ascending=True, inplace=True)
df_pred_validation.tail()
```

```python
min_score = round(df_pred_validation['Score'].min(),0)
max_score = round(df_pred_validation['Score'].max(),0)
print(f'The minimum score is: {min_score}')
print('')
print(f'The maximum score is: {max_score}')
```

# Statistical analysis

```python
def score_cat(score):
    if(score<100):
        return "0_100"
    if(score<200):
        return "100_200"
    if(score<300):
        return "200_300"
    if(score<400):
        return "300_400"
    if(score<500):
        return "400_500"
    if(score<600):
        return "500_600"
    if(score<700):
        return "600_700"
    if(score<800):
        return "700_800"
    if(score<900):
        return "800_900"
    else:
        return ">900"
    
df_pred_validation["score_cat"] = df_pred_validation.apply(lambda x: score_cat(x["Score"]), axis=1)
```

```python
def score_zones_cat(score):
    if(score<201):
        return "0_200"
    if(score<401):
        return "201_400"
    if(score<601):
        return "401_600"
    if(score<801):
        return "601_800"
    else:
        return "801_1000"
    
df_pred_validation["score_zones_cat"] = df_pred_validation.apply(lambda x: score_zones_cat(x["Score"]), axis=1)
```

```python
def cria_perc(df, vars_groupBy, var_count):
    df_perc = df.groupby(vars_groupBy).agg(cont = (var_count, 'count'))
    df_perc['aux'] = 1
    if type(vars_groupBy) == list:
        group_vars = ['aux']+[vars_groupBy[0]]
        df_perc_agg = df_perc.groupby(group_vars).agg(total = ('cont', 'sum'))
    else:
        group_vars = 'aux'
        df_perc_agg = df_perc.groupby(group_vars).agg(total = ('cont', 'sum'))
    
    df_perc = df_perc.reset_index()
    df_perc = df_perc.merge(df_perc_agg, on=group_vars, how='left')
    df_perc['percentual'] = 100*(df_perc['cont']/df_perc['total'])
  
    return df_perc

def plot_perc(df, varx=None, vary=None, hue=None, title=None):
    fig, ax = plt.subplots(figsize=(12,6))
    plot = sns.barplot(x=varx, y=vary, data=df, hue=hue)
    for p in plot.patches:
        perc = p.get_height()
        percentage = '{:.1f}%'.format(perc)
        x = p.get_x() + p.get_width()/2.
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', color='black', fontsize=8, textcoords='offset points',xytext=(0, 10))
    
    ax.set_ylim(0,109)
    plt.title(title)
    plt.show()
```
